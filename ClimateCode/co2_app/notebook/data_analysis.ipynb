```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_c02 = pd.read_csv('../utils/data/co2.csv')
print(df_c02.head())
```

```python
# Medidas descripitivas

# Cambiar de tipo objeto a int64
print(df_c02['pais'].unique())


# Conteo de todas las emsisiones percapita desde el anio 2000 al 2010
print(f"Miles de toneladas de CO2: {df_c02['co2_total'].sum()}") # Miles de toneladas
print(f"Toneladas de CO2 por habitante : {df_c02['co2_habitante'].sum()}") # Toneladas percapita
print(f"Toneladas de CO2 por cada millón de dólares de PIB: {df_c02['co2_producto_interno'].sum()}") # Toneladas 

#print(df_c02.dtypes)
```

```python
# Grafico de tendendia de emisiones de CO2 percapita en El Salvador

df_c02_el_salvador = df_c02[df_c02['pais'] == 'El Salvador']
print(df_c02_el_salvador)
plt.plot(df_c02_el_salvador['año'], df_c02_el_salvador['co2_habitante'])
plt.xlabel('Año')
plt.ylabel('CO2 per capita')
plt.title('Tendencia de emisiones de CO2 per capita en El Salvador')
plt.show()
```

```python
### Comparacion de emisiones de CO2 percapita en El Salvador y Guatemala desde 2004

df_c02_el_salvador = df_c02[(df_c02['pais'] == 'El Salvador') & (df_c02['año'] >= 2004)]
df_c02_guatemala = df_c02[(df_c02['pais'] == 'Guatemala') & (df_c02['año'] >= 2004)]

plt.plot(df_c02_el_salvador['año'], df_c02_el_salvador['co2_habitante'], label='El Salvador')

plt.plot(df_c02_guatemala['año'], df_c02_guatemala['co2_habitante'], label='Guatemala')
plt.xlabel('Año')
plt.ylabel('CO2 per capita')
plt.title('Comparacion de emisiones de CO2 per capita en El Salvador y Guatemala desde 2004')
plt.legend()
plt.show()
```

```python
### Comparacion de emisiones de CO2 con respecto al PIB en El Salvador y Guatemala desde 2004

df_c02_el_salvador = df_c02[(df_c02['pais'] == 'El Salvador') & (df_c02['año'] >= 2004)]
df_c02_guatemala = df_c02[(df_c02['pais'] == 'Guatemala') & (df_c02['año'] >= 2004)]

plt.plot(df_c02_el_salvador['año'], df_c02_el_salvador['co2_producto_interno'], label='El Salvador')

plt.plot(df_c02_guatemala['año'], df_c02_guatemala['co2_producto_interno'], label='Guatemala')
plt.xlabel('Año')
plt.ylabel('CO2 per capita')
plt.title('Comparacion de emisiones de CO2 con respecto al PIB en El Salvador y Guatemala desde 2004')
plt.legend()
plt.show()
```

```python
### Comparacion de mile de toneladas en emisiones de CO2 en El Salvador y Guatemala desde 2004

df_c02_el_salvador = df_c02[(df_c02['pais'] == 'El Salvador') & (df_c02['año'] >= 2004)]
df_c02_guatemala = df_c02[(df_c02['pais'] == 'Guatemala') & (df_c02['año'] >= 2004)]

plt.plot(df_c02_el_salvador['año'], df_c02_el_salvador['co2_total'], label='El Salvador')

plt.plot(df_c02_guatemala['año'], df_c02_guatemala['co2_total'], label='Guatemala')
plt.xlabel('Año')
plt.ylabel('CO2 per capita')
plt.title('Comparacion de mile de toneladas en emisiones de CO2 en El Salvador y Guatemala desde 2004')
plt.legend()
plt.show()
```
